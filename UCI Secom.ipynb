## 1. 데이터

### 1-1. 데이터 읽어오기

```python
# for basic operations
import numpy as np
import pandas as pd

# for visualizations
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("uci-secom.csv")
df
```

```python
# PASS/ FAIL 변수 확인
df['Pass/Fail'].value_counts().plot.pie(explode=[0, 0.1], autopct='%.1f%%')
```

```python
df['Pass/Fail'].value_counts()
```

```python
df_stat = df.describe().transpose()
```

```python
df_stat
```

### 1-2. 데이터 전처리

```python
# 피쳐 이름 센서로 변경
df_except = df.loc[:,['Time', 'Pass/Fail']] 
```

```python
df_change = df.drop(['Time', 'Pass/Fail'], axis=1).add_prefix('Sensor_')
```

```python
df = pd.concat([df_except, df_change], axis=1)
```

```python
df.head()
```

```python
df_stat
```

```python
# 무의미한 수치값인 열들 제거 (표준편차 == 0 인 컬럼들 제거)
df_stat[df_stat['std'] == 0]
# 116개의 무의미한 컬럼 검색
```

```python
remove_nm_cols = df_stat[df_stat['std'] == 0].index
remove_nm_cols
```

```python
df_second = df.drop(remove_nm_cols, axis=1) # 표준편차 0값이 컬럼들 제거
```
